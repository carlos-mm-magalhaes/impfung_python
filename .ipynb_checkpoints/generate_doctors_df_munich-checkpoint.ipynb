```python
import requests
import pandas as pd
from collections import namedtuple

BASE_URL = 'https://hausaerzte.site'
lines = []
Line = namedtuple('Line', 'doctor street PLZ email url')
```

```python
# iterate through all main pages 
for pag in range(1,36):
    url_inst = "https://hausaerzte.site/by/muenchen/"+str(pag)
    res = requests.get(url_inst)
    soup = bs4.BeautifulSoup(res.text,"lxml")
    items = soup.select('div.company')
    
    # iterate through all items in the page
    for i in items:
        # Get first information about the doctor
        doctor = i.select('span')[0].get_text() #name
        street = i.select('span')[1].get_text() #street
        plz = i.select('span')[2].get_text() #PLZ

        #print(i.select('a.link-overlay')[0].get('href'))
        url = BASE_URL + i.select('a.link-overlay')[0].get('href')
        print(url)
        res_arzt = requests.get(url)
        soup_arzt = bs4.BeautifulSoup(res_arzt.text,"lxml")

        # Get Email from the doctor
        try:
            email = soup_arzt.select('table.table.table-contact-data')[0].find(itemprop='email').get_text()
        except:
            email = ''

        lines.append(Line(doctor, street, plz, email, url))
```
